```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

msft = pd.read_csv('run-BERT1e-4v2-tag-loss.csv')

def filter(raw_data, beta=0.99):
    current_value = raw_data[0]
    for i in range(len(raw_data)):
        raw_data[i] = current_value * beta + (1 - beta) * raw_data[i]
        current_value = raw_data[i]
    return raw_data

def plot_files(files, names, y_name, title, save = ""):
    pandas_file = [pd.read_csv(f) for f in files]
    
    plt.figure()
    for p in pandas_file:
        np_value = np.array(p['Value'])
        np_value = filter(np_value)
        plt.plot(np_value)
        
    plt.legend(names)
    plt.ylabel(y_name)
    plt.xlabel('Step')
    plt.title(title)
    plt.grid(True)
    ax = plt.gca()
    if save != "":
        plt.savefig(save+".png")
    plt.show()
    
files = ["run-20200611-095409-tag-acc_1.csv", "run-IMPORTANT_REsULTS_BERT1e-5FULL_GOOD_RES-tag-acc_1.csv"]
names = ["Our model", "Classic Bert"]
plot_files(files, names, "Accuracy", "Accuracy start short", "full_train_acc1")

files = ["run-20200611-095409-tag-acc_2.csv", "run-IMPORTANT_REsULTS_BERT1e-5FULL_GOOD_RES-tag-acc_2.csv"]
names = ["Our model", "Classic Bert"]
plot_files(files, names, "Accuracy", "Accuracy end short", "full_train_acc2")

files = ["run-20200611-095409-tag-acc_3.csv", "run-IMPORTANT_REsULTS_BERT1e-5FULL_GOOD_RES-tag-acc_3.csv"]
names = ["Our model", "Classic Bert"]
plot_files(files, names, "Accuracy", "Accuracy long", "full_train_acc2")


#other models 
files = ["run-ALBERT_ALBERT_FULL1e-5-tag-acc_1.csv", "run-BERT1e-4v2-tag-acc_1.csv", 
         "run-IMPORTANT_REsULTS_BERT2EPOCHS1e-5-tag-acc_1.csv"]
names = ["ALBERT", "BERT 1e-4", "BERT1e-5"]
plot_files(files, names, "Accuracy", "Accuracy start short", "HP_acc1")

files = ["run-ALBERT_ALBERT_FULL1e-5-tag-acc_2.csv", "run-BERT1e-4v2-tag-acc_2.csv", 
         "run-IMPORTANT_REsULTS_BERT2EPOCHS1e-5-tag-acc_2.csv"]
names = ["ALBERT", "BERT 1e-4", "BERT1e-5"]
plot_files(files, names, "Accuracy", "Accuracy end short", "HP_acc2")


files = ["run-ALBERT_ALBERT_FULL1e-5-tag-acc_3.csv", "run-BERT1e-4v2-tag-acc_3.csv", 
         "run-IMPORTANT_REsULTS_BERT2EPOCHS1e-5-tag-acc_3.csv"]
names = ["ALBERT", "BERT 1e-4", "BERT1e-5"]
plot_files(files, names, "Accuracy", "Accuracy start long", "HP_acc3")

files = ["run-ALBERT_ALBERT_FULL1e-5-tag-loss.csv", "run-BERT1e-4v2-tag-loss.csv", 
         "run-IMPORTANT_REsULTS_BERT1e-5FULL_GOOD_RES-tag-loss.csv"]
names = ["ALBERT", "BERT 1e-4", "BERT1e-5"]
plot_files(files, names, "Loss", "Loss", "HP_loss")


```
